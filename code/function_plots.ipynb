```julia
using Plots, LaTeXStrings, ComplexPhasePortrait, SpecialFunctions
```

# Create a grid of complex numbers in $[-10,10]\times i[-10,10]$

```julia
δ = 0.01
t = -10:δ:10
z = t' .+ reverse(im*t)
```

# Plots

```julia
heatmap(t,t,real.(z),xlabel=L"Re\{z\}",ylabel=L"Im\{z\}")
```

```julia
heatmap(t,t,angle.(z.^2),xlabel=L"Re\{z\}",ylabel=L"Im\{z\}")
```

```julia
heatmap(t,t,angle.(z.^2),xlabel=L"Re\{z\}",ylabel=L"Im\{z\}")
```

# Phase plots of complex functions

```julia
phaseplot(-1:0.011:1,-1:0.011:1, z->z^-2, aspect_ratio=1)
```
