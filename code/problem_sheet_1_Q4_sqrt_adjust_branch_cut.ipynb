```julia
δ = 0.01
t = -1:δ:1
z = t' .+ im*(t);
```

```julia
heatmap(t,t,angle.(sqrt.(z)),xlabel="Real",ylabel="Imag")
```

Now make our own. The logic here is that if $\arg z \in [\theta,\pi]$, then we want to be on a different branch to the standard square root function `sqrt` (which has its branch cut along $\arg z=\pi$). We can switch branches simply by multiplying by $-1$. Otherwise, our function should match the standard square root function `sqrt` .

```julia
function sqrt_modified(z, θ)
   if (θ <= angle(z))
        return -sqrt(z) #switch branch
    else
        return sqrt(z) #dont switch branch
    end
end
```

```julia
Z = sqrt_modified.(z,π-0.5)
B = sqrt_modify.(z, π-0.5)
heatmap(t,t,angle.(sqrt_modified.(z,π-0.5)),xlabel="Real",ylabel="Imag")
```
